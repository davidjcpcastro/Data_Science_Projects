### Question 1

1.
    * `TRANSFORMATION APPLIED: Creation of 'listings' Collection`
    * `REASONING BEHIND: The initial collection 'listingsAndReviews_new' has a lot of different information, since the fields regarding the listings have the information that will be queried the most, it makes sense to transfer it to a new collection. `
    * `EXPECTED RESULT: Using this transformation queries will run faster and since they will have to retrieve less fields`

2.
    * `TRANSFORMATION APPLIED: Application of Attribute Pattern`
    * `REASONING BEHIND: Since there were 6 fields that were common to all documents in the 'listings' collection it made sense to apply an attribute pattern to 'review_scores'`
    * `EXPECTED RESULT: Using this transformation the attributes will be optimized `

```python
# LISTINGS
project_operation = {
    "$project" : {"_id": 1,
                        "access": "$access",
                        "accommodates": "$accommodates",
                        "address": "$address",
                        "amenities": "$amenities",
                        "availability": "$availability",
                        "bathrooms": "$bathrooms",
                        "bed_type": "$bed_type",
                        "bedrooms": "$bedrooms",
                        "beds": "$beds",
                        "calendar_last_scraped":"$calendar_last_scraped",
                        "cancellation_fee":"$cancellation_fee",
                        "description":"$description",
                        "extra_people":"$extra_people",
                        "first_review":"$first_review",
                        "guests_included":"$guests_included",
                        "host_id": { "$toInt": "$host_id" },
                        "house_rules":"$house_rules",
                        "images":"$images",
                        "interaction":"$interaction",
                        "last_scraped":"$last_scraped",
                        "listing_url":"$listing_url",
                        "maximum_nights":"$maximum_nights",
                        "minimum_nights":"$minimum_nights",
                        "name":"$name",
                        "neighbourhood_overview":"$neighbourhood_overview",
                        "notes":"$notes",
                        "number_of_reviews":"$number_of_reviews",
                        "price":"$price",
                        "property_type":"$property_type",
                        "review_scores": {
                            "checkin":"$review_scores_checkin",
                            "cleanliness":"$review_scores_cleanliness",
                            "communication":"$review_scores_communication",
                            "location":"$review_scores_location",
                            "rating":"$review_scores_rating",
                            "value":"$review_scores_value"
                        },
                        "room_type":"$room_type",
                        "security_deposit":"$security_deposit",
                        "space":"$space",
                        "summary":"$summary",
                        "transactions":"$transactions",
                        "transit":"$transit",
                        "last_review":"$last_review"
                    }
}

merge_operation = { 
        "$merge" : {
            "into" : "listings",
            "whenMatched": "replace"
        } 
    }
pipeline = [project_operation,merge_operation]

result = db.listingsAndReviews_new.aggregate(pipeline)
```

```python
#OBJECTIVE: Transform the price inside transactions.transactions in a float instead of a string

# iterate through each document in the listings collection
for listing in db.listings.find():
    # access the transactions subdocument for the current listing
    transactions = listing['transactions']['transactions']

    # iterate through each transaction subdocument in the transactions list
    for transaction in transactions:
        # convert the price field from string to float
        transaction['price'] = float(transaction['price'])

    # update the transactions subdocument with the modified transactions list
    db.listings.update_one(
        {'_id': listing['_id']},
        {'$set': {'transactions.transactions': transactions}}
    )
```

```python
# Split 'Washer / Dryer' into 2 different amenities
value_to_split = 'Washer / Dryer'
new_values = ['Washer', 'Dryer']

# Update the documents in the collection to split the specified value in the amenities field
result = db.listings.update_many(
    {"amenities": value_to_split},
    [
        {"$set": {
            "amenities": {
                "$concatArrays": [
                    [
                        {"$arrayElemAt": ["$amenities", {"$indexOfArray": ["$amenities", value_to_split]}]},
                        new_values,
                        {"$slice": ["$amenities", {"$add": [{"$indexOfArray": ["$amenities", value_to_split]}, 1]}, {"$size": "$amenities"}]}
                    ]
                ]
            }
        }}
    ]
)

# Print the number of documents updated
print(f"{result.modified_count} documents updated")
```

```python
# Change the amenity Pool to Swimming pool
# Define the old and new amenity names
old_amenity = 'Pool'
new_amenity = 'Swimming pool'

# Update the documents in the collection to replace the old amenity name with the new one
result = db.listings.update_many(
    {"amenities": old_amenity},
    [
        {"$set": {
            "amenities": {
                "$map": {
                    "input": "$amenities",
                    "as": "amenity",
                    "in": {
                        "$cond": [
                            {"$eq": ["$$amenity", old_amenity]},
                            new_amenity,
                            "$$amenity"
                        ]
                    }
                }
            }
        }}
    ]
)

# Print the number of documents updated
print(f"{result.modified_count} documents updated")
```

```python
pipeline = [
    # Select only listings with valid values for 'maximum_nights', 'minimum_nights' and 'address.market' which is the city
    {
        "$match": {
            "maximum_nights": {"$exists": True, "$ne": None, "$ne": ''},
            "minimum_nights": {"$exists": True, "$ne": None, "$ne": ''},
            "address.market": {"$exists": True, "$ne": None, "$ne": ''}
        }
    },
    {
        "$project": {
            "maximum_nights_int": {"$toInt": "$maximum_nights"},
            "minimum_nights_int": {"$toInt": "$minimum_nights"},
            "property_type":"$property_type",
            "city":"$address.market"
        }
    },
    {
        "$project": {
            "duration_stay": {
                "$subtract": ["$maximum_nights_int", "$minimum_nights_int"]
            },
            "property_type":"$property_type",
            "city":"$city"
        }
    },
    {
        "$group": {
        "_id": {'city': '$city', 'property_type': "$property_type"},
        "average_duration_stay": {"$avg": "$duration_stay"}
        }
    },
    {
        '$project':{
            '_id':0,
            'City':'$_id.city',
            'Property type':'$_id.property_type',
            'Average duration stay':'$average_duration_stay'
        }
    },
    {
        "$sort":{
            '_id.city':1
        }
    },

    # Create new collection to store information about the duration of stays
    {
        '$out':'duration_stays'
    }
]


result = db.listings.aggregate(pipeline)

for doc in result:
    pprint(doc)
```

Question: For each property we store review scores across different metrics (accuracy, check-in, cleanliness etc). We want to be able to easily query the average score across all of these metrics, including any new metrics that might be added without changing the query. Adjust the data model so this can be done and show the query for an example property.

10.
    * `TRANSFORMATION APPLIED: Creation of 'metrics' Collection. match stage that filters out documents that do not have all the required fields, project stage that reshapes the documents and creates new fields for each metric of interest, addFields stage that calculates the average of the metrics and adds it as a new field, out stage that writes the output to a new collection.`
    * `REASONING BEHIND: This pipeline applies transformations to the data to prepare it for analysis, calculating the average of different metrics that could be used to evaluate the quality of Airbnb listings.  Restructuring the data model creating a new collection for the review metrics and store a reference to the corresponding listing in each metric document, allow us to address the problem of being able to easily query the average score across all of the review metrics, including any new metrics that might be added in the future.The match stage is necessary to remove documents that do not have all the required fields, while the project stage creates new fields that are easier to work with and renames existing fields to uppercase. The addFields stage is used to calculate the average of the metrics and add it as a new field, and the out stage writes the output of the pipeline to a new collection for further analysis. `
    * `EXPECTED RESULT: The expected result of running the queries given this transformation is a new collection named "metrics" that contains documents with the name, average metrics, and ID fields for each Airbnb listing that had all the required fields. This collection will make it easier to analyze the quality of the listings and identify any patterns or trends in the data, as well as easily adding more metrics in the future.`

```python
pipeline = [
    {
        "$match": {
            # only select documents that have the required fields
            "review_scores.checkin": {"$exists": True},
            "review_scores.cleanliness": {"$exists": True},
            "review_scores.communication": {"$exists": True},
            "review_scores.location": {"$exists": True},
            "review_scores.rating": {"$exists": True},
            "review_scores.value": {"$exists": True},
        }
    },
    {
        # The $project stage is used to reshape the documents in the collection. 
        # In this stage, we are creating new fields for each of the metrics we are interested in
        "$project": {
            "_id": 0,
            "NAME": "$name",
            "CHECKIN": "$review_scores.checkin",
            "CLEANLINESS": "$review_scores.cleanliness",
            "COMMUNICATION": "$review_scores.communication",
            "LOCATION": "$review_scores.location",
            "RATING": {"$multiply": ["$review_scores.rating", 0.1]},  # divide by 10
            "VALUE": "$review_scores.value",
        }
    },
    {
        "$addFields": {
            # calculate the average of the metrics for each document
            "AVG_METRICS": {
                "$round": [   #round up to 3 decimal numbers
                    {"$avg": ["$CHECKIN", "$CLEANLINESS", "$COMMUNICATION", "$LOCATION", "$RATING", "$VALUE"]},
                    3
                ]
            },
            # add a reference to the original document's _id value
            "ORIGINAL_ID": "$_id"
        }
    },
    {
        "$out": "metrics" # write the output of the aggregation pipeline to a new collection "metrics".
    }
]

# execute the aggregation pipeline and insert the result into the metrics collection
metrics = db["metrics"]
result = list(db.listings.aggregate(pipeline))
if result:
    # capitalize the field names before inserting into the metrics collection
    result[0] = {k.upper(): v for k, v in result[0].items()}
    metrics.insert_one(result[0])
```

10.
    * `TRANSFORMATION APPLIED: We added a new review metric called X_FACTOR with a score of 10 to the new document created in the question 13, which has the id = 9993191. This was firstly updated in the collection 'listings'. Then to compute the average between all review metrics, including the X_FACTOR one, we applied the same query that we used in the question 10, which is a reference to the collection listings, creating then a new document in the collection 'metrics' with all this metrics and its average.`
    * `REASONING BEHIND: By doing this we are easily calculating the average of different metrics that could be used to evaluate the quality of Airbnb listings.  Restructuring the data model creating a new collection for the review metrics and store a reference to the corresponding listing in each metric document, allow us to address the problem of being able to easily query the average score across all of the review metrics, including this new metric that was added (X_FACTOR). `
    * `EXPECTED RESULT: The expected result is a new metric called X_FACTOR with score of 10 in the new document created with the id = 9993191. As well as the average of all metrics of this new document in another new document of the collection "metrics" by using a reference to make it easier to analyze the quality of each property.`

```python
# update the new document with the new review metric and score
db.listings.update_one({"_id": '9993191'}, {"$set": {"review_scores.X_FACTOR": 10}})

# if we want to confirm the step above just uncomment the folling lines

#cursor = db.listings.find({"_id": "9993191"})
#for document in cursor:
#    pprint(document)
```

```python
pipeline = [
    {
        "$match": {
            # match to the new document created
            '_id': '9993191',
            # only select documents that have the required fields
            "review_scores.checkin": {"$exists": True},
            "review_scores.cleanliness": {"$exists": True},
            "review_scores.communication": {"$exists": True},
            "review_scores.location": {"$exists": True},
            "review_scores.rating": {"$exists": True},
            "review_scores.value": {"$exists": True},
            "review_scores.X_FACTOR": {"$exists": True},
        }
    },
    {
        # The $project stage is used to reshape the documents in the collection. 
        # In this stage, we are creating new fields for each of the metrics we are interested in
        "$project": {
            "_id": 1,
            "NAME": "$name",
            "CHECKIN": "$review_scores.checkin",
            "CLEANLINESS": "$review_scores.cleanliness",
            "COMMUNICATION": "$review_scores.communication",
            "LOCATION": "$review_scores.location",
            "RATING": {"$multiply": ["$review_scores.rating", 0.1]},  # divide by 10
            "VALUE": "$review_scores.value",
            "X_FACTOR": "$review_scores.X_FACTOR",
        }
    },
    {
        "$addFields": {
            # calculate the average of the metrics for each document
            "AVG_METRICS": {
                "$round": [   #round up to 3 decimal numbers
                    {"$avg": ["$CHECKIN", "$CLEANLINESS", "$COMMUNICATION", "$LOCATION", "$RATING", "$VALUE", "$X_FACTOR"]},
                    3
                ]
            }
        }
    },
    {
        # output the results to a new collection named "metrics"
        "$out": "metrics"
    }
]

# execute the aggregation pipeline
db.listings.aggregate(pipeline)
```
